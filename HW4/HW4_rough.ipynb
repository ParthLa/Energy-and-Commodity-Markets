```python
sheets = ["Bitcoin.csv", "Copper.csv", "Oil.csv", "Silver.csv"]
# Take bitcoin index and put it on all assets
df_returns = []
for sheet in sheets:
    df = pd.read_csv(sheet)[1:]
    for col in df.columns:
        try:
            df[col] = df[col].astype(float)
        except Exception as e:
            pass
    df["Ret"] = np.log(df["Close"]/df["Close"].shift(1))
    print(len(df["Ret"]))
    df_returns.append(df["Ret"])
    print(len(df_returns))
# df_returns = np.array(df_returns)
```
